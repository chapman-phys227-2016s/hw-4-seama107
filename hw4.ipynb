The Global Historical Climatology network publishes a daily database that includes daily readings for climate records all across the globe (although this set is just limited to California). The database contains information on up to 40 different climatal elements including information on rainfall, snowfall, wind, clouds, temperature and more. We will be doing an analysis on a single station's data for rainfall, so first we have to get the data and pick a station.


```python
#pd.set_option('display.max_colwidth',100)
#np.set_printoptions(threshold=np.nan)
data = pd.DataFrame(pd.read_csv('598354.csv'))
print "There are " + str(len(data["STATION_NAME"].unique())) + " unique stations in the data base."
print
print
print data["STATION_NAME"].unique()
```

That is a lot of stations! I'm going to pick "ESCONDIDO 7.2 SE CA US" because it's my hometown.

Let's look at the data collected by the Escondido station and what we can learn from it.

```python
chosen_station = "ESCONDIDO 7.2 SE CA US"
escondido_data = data[data.STATION_NAME == chosen_station]
escondido_data
```

Unfortunately it looks like most of the data fields besides "PRCP" are not filled out. Let's verify this.

```python
def print_if_unique_data(data, chosen_station, category):
    if len(data[data.STATION_NAME == chosen_station][category].unique()) == 1:
        print "No station information on " + category
    else:
        print "Information found about " + category

print_if_unique_data(data, chosen_station, "MDPR")
print_if_unique_data(data, chosen_station, "DAPR")
print_if_unique_data(data, chosen_station, "TMIN")
print_if_unique_data(data, chosen_station, "TMAX")
print_if_unique_data(data, chosen_station, "TOBS")
print_if_unique_data(data, chosen_station, "PRCP")

```

So it seems to be the case that there is no data on the following data types:
    - Multi-day Precipitation
    - How long the multi-day Precipitation collection lasted
    - Minimum temperature
    - Maximum temperature
    - Temperature at the time of the observation

However, the Precipitation category does seem to be filled out so let's glance at that and see if we can at least learn the minimum and maximum rainfall that Escondido has had since 2014

```python
minimum_PRCP = min(escondido_data["PRCP"])
maximum_PRCP = max(escondido_data["PRCP"])
average_PRCP = np.sum(escondido_data["PRCP"])/(float(len(escondido_data["PRCP"])))
print "Minimum rainfall: " + str(minimum_PRCP) + " tenths of a mm."
print "Maximum rainfall: " + str(maximum_PRCP) + " tenths of a mm."
print "Average daily rainfall: " + str(average_PRCP) + " x 10^-4 m."


print "The the max rainfall occured on: "
escondido_data[data.PRCP == maximum_PRCP]["DATE"]
```

As we can see there is very very little rain in Escondido, as with a lot of San Diego county. We can see that the maximum rain fall of 3.2 cm occured on December 4th of 2014, which makes sense, being in the middle of winter. Let's see if we can visualize the data of rainfall to get a more intuitive picture of it.

```python
xList = np.arange(len(escondido_data))
yList = escondido_data["PRCP"]
plt.plot(xList, yList)
plt.xticks(np.arange(0, len(xList), 60))
plt.show()
```

Above we can see the rainfall amount plotted against the number of days since the data collection started. We can clearly see the seasonal trend; most of the rainfall is clouded around the beginnings of the years (x = 0 and x = 364), corresponding to winter. As we move away from those points, the rainfall stops. For example, looking at just the month of june, we should hardly see any rain at all.

```python
escondido_june_data = escondido_data[data.DATE > 20150600]
escondido_june_data = escondido_june_data[data.DATE < 20150700]
escondido_june_data
```
